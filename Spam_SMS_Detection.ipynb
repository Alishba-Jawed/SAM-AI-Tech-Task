```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/spam.csv", encoding = "Latin-1")
```

```python
data
```

```python
time_counts = data["v1"].value_counts()
time_counts.values
```

```python
data = data[["v1", "v2"]]
```

```python
data.columns= ["Labels","Messages"]
```

```python
data["Labels"].unique()
```

Categorical to Numerical by Mapping

```python
data = data.copy()# Warning means Pandas is unsure if you're modifying original data or a copy (.copy() to avoid it)
data["Labels"] = data["Labels"].map({"spam" : 1, "ham" : 0})
```

Splitting For Training


```python
X = data[["Messages"]]
```

```python
y = data["Labels"]
```

```python
X
```

```python
y
```

Training

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    data["Messages"], data["Labels"],  test_size= 0.30, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

Categorical to Numerical Using TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
```

```python
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
model.fit(X_train_vec, y_train)
```

Testing


```python
y_train_predict = model.predict(X_train_vec)
y_test_predict = model.predict(X_test_vec)
```

```python
from sklearn.metrics import accuracy_score
```

```python
Training_Accuracy = accuracy_score(y_train, y_train_predict)
Training_Accuracy
```

```python
Testing_Accuracy = accuracy_score(y_test_predict, y_test)
Testing_Accuracy
```

```python
print(f"Training accuracy : {Training_Accuracy}\nTest accuracy : {Testing_Accuracy}")
```
